```python
from nltk.corpus import stopwords
```

```python
worf_quote = "Sir, I protest. I am not a merry man!"
```

```python
worf_token = word_tokenize(worf_quote)
```

```python
worf_token
```

```python
stop_word = set(stopwords.words("english"))   #You have a list of the words in worf_quote, so the next step is to create a set of stop words to filter words_in_quote. For this example, you’ll need to focus on stop words in "english
```

```python
filtered_list = []
```

```python
#You used .casefold() on word so you could ignore whether the letters in word were uppercase or lowercase.

filtered_list = [
    word for word in worf_token if word.casefold() not in stop_word
]
```

```python
filtered_list
```

```python
#Stemming is a text processing task in which you reduce words to their root, which is the core part of a word. For example, the words “helping” and “helper” share the root “help.”
from nltk.stem import PorterStemmer
```

```python
stemmer = PorterStemmer()
```

```python
string_for_stemming = "The crew of the USS Discovery discovered many discoveries. Discovering is what explorers do."
```

```python
words = word_tokenize(string_for_stemming)
```

```python
words
```

```python
stemmed_words = [stemmer.stem(word) for word in words]
```

```python
from nltk import WordNetLemmatizer
```
